```python
# Import Dependencies
import pandas as pd
import hvplot.pandas
from pathlib import Path
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
# Make a reference to the comic_books_expanded.csv file path
csv_path = "Data/data.csv"

# Import the data.csv file as a DataFrame
scale_df = pd.read_csv(csv_path, encoding="utf-8",
                   sep = ';',
                   engine = 'python')

scale_df.head()
```

```python
# Make a reference to the comic_books_expanded.csv file path
csv_path = "Data/modified_data.csv"

# Import the data.csv file as a DataFrame
one_hot_df = pd.read_csv(csv_path, encoding="utf-8",
                   sep = ',',
                   engine = 'python')

one_hot_df.head()
```

```python
# Scaling the numeric columns
school_data_scaled = StandardScaler().fit_transform(scale_df[["Application order", "Previous qualification (grade)", "Admission grade","Age at enrollment","Curricular units 1st sem (credited)","Curricular units 1st sem (enrolled)","Curricular units 1st sem (evaluations)","Curricular units 1st sem (approved)","Curricular units 1st sem (grade)","Curricular units 1st sem (without evaluations)","Curricular units 2nd sem (credited)","Curricular units 2nd sem (enrolled)","Curricular units 2nd sem (evaluations)","Curricular units 2nd sem (approved)","Curricular units 2nd sem (grade)","Curricular units 2nd sem (without evaluations)","Unemployment rate","Unemployment rate","GDP"]])

# Creating a DataFrame with the scaled data
df_school_transformed = pd.DataFrame(school_data_scaled, columns=["Application order", "Previous qualification (grade)", "Admission grade","Age at enrollment","Curricular units 1st sem (credited)","Curricular units 1st sem (enrolled)","Curricular units 1st sem (evaluations)","Curricular units 1st sem (approved)","Curricular units 1st sem (grade)","Curricular units 1st sem (without evaluations)","Curricular units 2nd sem (credited)","Curricular units 2nd sem (enrolled)","Curricular units 2nd sem (evaluations)","Curricular units 2nd sem (approved)","Curricular units 2nd sem (grade)","Curricular units 2nd sem (without evaluations)","Unemployment rate","Unemployment rate","GDP"])

# Display sample data
df_school_transformed.head()
```

```python
columns = ["Marital status", "Application mode", "Course", "Daytime/evening attendance\t","Previous qualification","Nacionality","Mother's qualification","Father's qualification","Mother's occupation","Father's occupation","Displaced","Educational special needs","Debtor","Tuition fees up to date","Gender","Scholarship holder","International","Target"]

finished_df  = df_school_transformed.copy()
for i in columns:

    # Perform one-hot encoding on column 'A'
    one_hot = pd.get_dummies(one_hot_df[i], prefix=i)
    # Convert boolean values to integers (0 and 1)
    one_hot = one_hot.astype(int)

    # Concatenate the one-hot encoded columns with the original DataFrame
    finished_df = pd.concat([finished_df, one_hot], axis=1)
finished_df
```

```python
finished_df.to_csv('Data/scaled_data.csv', index=False)
```

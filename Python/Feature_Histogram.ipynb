```python
# Import required libraries
import pandas as pd
import hvplot.pandas
import matplotlib.pyplot as plt
```

```python
# Make a reference to the comic_books_expanded.csv file path
csv_path = "../Data/data.csv"

# Import the data.csv file as a DataFrame
df = pd.read_csv(csv_path, encoding="utf-8",
                   sep = ';',
                   engine = 'python')

df.head(5)
```

```python
#Drop the Target column
df=df.drop(columns=['Target'])
```

```python
#Rename one colomn title
df=df.rename(columns={'Daytime/evening attendance\t':'Daytime evening attendance'})
```

```python
#Check title of all columns
all_columns=df.columns.values
all_columns
```

```python
#Create a loop to make histogram for all feature columns
for i in all_columns:
    df.hist(column=i)
plt.savefig('../Figure/{}.png')
```
